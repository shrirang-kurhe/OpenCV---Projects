```python
import cv2

# Load the Haarcascade
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Initialize the camera
cap = cv2.VideoCapture(0)
if not cap.isOpened():
    print("Error: Could not open the camera.")
    exit()

while cap.isOpened():
    has_frame, frame = cap.read()
    if not has_frame:
        print("Error: Frame not captured.")
        break

    # Convert to grayscale
    imag = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect faces
    faces = face_cascade.detectMultiScale(imag, 1.1, 4)

    # Draw rectangles around detected faces
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 0, 0), 3)

    # Display the frame
    cv2.imshow('image', frame)

    # Break the loop on 'q' key press
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release resources
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2           # Face Detection 

# Load the Haarcascade
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_russian_plate_number.xml')

# Initialize the camera
cap = cv2.VideoCapture(0)
if not cap.isOpened():
    print("Error: Could not open the camera.")
    exit()

while cap.isOpened():
    has_frame, frame = cap.read()
    if not has_frame:
        print("Error: Frame not captured.")
        break

    # Convert to grayscale
    imag = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect objects (e.g., license plates)
    faces = face_cascade.detectMultiScale(imag, 1.1, 4)

    # Draw rectangles around detected objects
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 0, 0), 3)

    # Display the frame
    cv2.imshow('image', frame)

    # Break the loop on 'q' key press
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release resources
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import os

# Load the Haarcascade
cascade_path = cv2.data.haarcascades + 'haarcascade_russian_plate_number.xml'
if not os.path.exists(cascade_path):
    print("Error: Haarcascade file not found.")
    exit()

face_cascade = cv2.CascadeClassifier(cascade_path)
if face_cascade.empty():
    print("Error: Failed to load Haarcascade file.")
    exit()

# Initialize the camera
cap = cv2.VideoCapture(1)
if not cap.isOpened():
    print("Error: Could not open the camera.")
    exit()

while cap.isOpened():
    has_frame, frame = cap.read()
    if not has_frame:
        print("Error: Frame not captured.")
        break

    # Convert to grayscale
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect objects
    faces = face_cascade.detectMultiScale(gray, 1.1, 4)

    # Draw rectangles around detected objects
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 0, 0), 3)

    # Display the frame
    cv2.imshow('image', frame)

    # Break the loop on 'q' key press
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release resources
cap.release()
cv2.destroyAllWindows()
```
