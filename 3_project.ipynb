### read the image using *OpenCV package*

```python
import cv2
image = cv2.imread(r"C:\Users\Shree\OneDrive\Desktop\Deroman image.jpg")
plt.imshow(image)
```

### Crop the Image

```python
import cv2
import matplotlib.pyplot as plt
image = cv2.imread(r"C:\Users\Shree\OneDrive\Desktop\Deroman image.jpg")

original_RGB = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
opencv_BGR = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.subplot(1,3,1).imshow(original_RGB)
plt.subplot(1,3,2).imshow(opencv_BGR)
plt.subplot(1,3,3).imshow(image)
```
